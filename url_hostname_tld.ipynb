```python
import csv
```

```python

f = open('openphish.csv','r')
rdr=csv.reader(f)
```

```python

from urllib import parse
import tldextract
```

```python


for i in rdr:
    print(i)
    url = parse.urlparse(i[0])
    hostname=url.netloc
    print(hostname)
    tld=tldextract.extract(i[0]).suffix
    print(tld)
```
